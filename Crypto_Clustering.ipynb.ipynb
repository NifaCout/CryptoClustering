```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file

scaler = StandardScaler()

selected_columns = df_market_data[['price_change_percentage_24h', 
                                   'price_change_percentage_7d', 
                                   'price_change_percentage_14d', 
                                   'price_change_percentage_30d', 
                                   'price_change_percentage_60d', 
                                   'price_change_percentage_200d', 
                                   'price_change_percentage_1y']]

scaled_data = scaler.fit_transform(selected_columns)
```

```python
# Create a DataFrame with the scaled data
# Set the coinid column as index
coin_id_index = df_market_data.index
df_scaled_data = pd.DataFrame(scaled_data, columns=['price_change_percentage_24h', 
                                                    'price_change_percentage_7d', 
                                                    'price_change_percentage_14d', 
                                                    'price_change_percentage_30d', 
                                                    'price_change_percentage_60d', 
                                                    'price_change_percentage_200d', 
                                                    'price_change_percentage_1y'], 
                                                    index=coin_id_index)

# Copy the crypto names from the original data
crypto_names = df_market_data.index.tolist()




# Display sample data
df_scaled_data.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []


# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k_values:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_scaled_data)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k_values
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** the best value for k would be 3 as it is the point where the inertia starts to decrease more slowly, indicating a point of diminishing returns. However, k=4 might also be a reasonable choice, depending on the specific context and goals of our analysis

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# Initialize the K-Means model using the best value for k
kmeans_model = KMeans(n_clusters=3, random_state=1)
```

```python
# Fit the K-Means model using the scaled data
kmeans_model.fit(df_scaled_data)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled data
predicted_clusters = kmeans_model.predict(df_scaled_data)


# Print the resulting array of cluster values.
print(predicted_clusters)
```

```python
# Create a copy of the DataFrame
df_market_data_copy = df_market_data.copy()
```

```python
# Add a new column to the DataFrame with the predicted clusters
df_market_data_copy['Cluster'] = predicted_clusters

df_market_data_copy['Cluster'] = df_market_data_copy['Cluster'].replace({
    0: 'Top Performers', 
    1: 'Average Performers', 
    2: 'Underperformers'
})


# Display sample data
df_market_data_copy.head(10)
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.

df_market_data_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    hover_cols='coin_id', 
    by='Cluster'
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca_model = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_model.fit(df_scaled_data)
crypto_pca = pca_model.transform(df_scaled_data)

# View the first five rows of the DataFrame. 
df_crypto_pca = pd.DataFrame(
    data=crypto_pca, 
    columns=['PC1', 'PC2', 'PC3'], 
    index=df_scaled_data.index
)
print(df_crypto_pca.head())
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
explained_variance = pca_model.explained_variance_ratio_
print(explained_variance)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The three main components sum up to 89.50% of the total data variation, showing that we've kept most of the important information even after reducing the data's complexity.

```python
# Create a new DataFrame with the PCA data.
df_pca = pd.DataFrame(
    data=crypto_pca, 
    columns=['PC1', 'PC2', 'PC3']
)

# Copy the crypto names from the original data
df_pca['coin_id'] = df_market_data.index.values

# Set the coinid column as index
df_pca.set_index('coin_id', inplace=True)

# Display sample data
df_pca.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    kmeans_model = KMeans(n_clusters=k, random_state=1)
    kmeans_model.fit(df_pca)
    inertia.append(kmeans_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    'k': k_values,  
    'inertia': inertia  
}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x='k', 
    y='inertia', 
    title='Elbow Curve', 
    xlabel='Number of clusters (k)', 
    ylabel='Inertia'
)
```

  * **Answer:**the best value for k in this case would likely be 3, as it's the first bend. But again, it would be better to experiment with 4 as well to see if they provide any additional meaningful information.


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** The original data you also found that the best value for k is 3, so it does not differ

### Visualize and Compare the Results

In this section, you will visually analyze the cluster analysis results by contrasting the outcome with and without using the optimization techniques.

```python
# Composite plot to contrast the Elbow curves
def calculate_inertia(data, k_range):
    inertia = []
    for i in k_range:
        kmeans = KMeans(n_clusters=i, random_state=1)
        kmeans.fit(data)
        inertia.append(kmeans.inertia_)
    return inertia
# Create a list of k values
k_values = list(range(1, 12))

# Calculate inertia for original scaled data and PCA data
inertia_original = calculate_inertia(df_scaled_data, k_values)
inertia_pca = calculate_inertia(df_pca, k_values)

# Create a DataFrame to store the results
df_elbow = pd.DataFrame({'k': k_values, 
                         'inertia_original': inertia_original, 
                         'inertia_pca': inertia_pca})

# Plot the results using hvPlot
plot = df_elbow.hvplot.line(
    x='k', 
    y=['inertia_original', 'inertia_pca'], 
    title='Elbow Curve Comparison', 
    xticks=k_values
)
plot
```

```python
# Composite plot to contrast the clusters

scatter_original = df_market_data_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    by="Cluster", 
    hover_cols=['coin_id'],
    title="Clusters Based on Original Data"
)

scatter_pca = df_pca_copy.hvplot.scatter(
    x="PC1", 
    y="PC2", 
    by="Cluster", 
    hover_cols=['coin_id'],
    title="Clusters Based on PCA Data"
)

# Now, create a composite plot to contrast the clusters
composite_plot = scatter_original + scatter_pca
composite_plot
```
